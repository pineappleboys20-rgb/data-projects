```python
df["date"] = pd.to_datetime(df["date"])

df["year"] = df["date"].dt.year
df["month"] = df["date"].dt.month
```

In this code we're making the date column a datetime object so we can extract the year and month from it

Now create a function that shows the highest, lowest, average  price of gold

```python
def basic_stats(df):
    highest = df['gold'].max()
    lowest = df['gold'].min()
    average_price = df['gold'].mean()
    lowest_date = df[df["gold"] == lowest]["date"].dt.strftime("%Y-%m-%d").values[0] #dt.strftime just formats the date into YYYY-MM-DD since before it was showing the time as well.
    highest_date = df[df["gold"] == highest]["date"].dt.strftime("%Y-%m-%d").values[0]
    
    print(f"Highest Gold price: ${highest:.2f} on {highest_date}")  
    print(f"Lowest Gold price: ${lowest:.2f} on {lowest_date}")     
    print(f"Average Gold price: ${average_price:.2f}")  
```

Here I analyze the highest, lowest and average gold price 
across 20 years of data. This gives us a baseline understanding 
of how gold has performed historically. Next create a function that groups the price by yearly

```python
def yearly_performance (df):
    grouped_by_year = df.groupby('year') ['gold'].mean()
    
    for year in grouped_by_year:
        print(f"{year}: ${grouped_by_year[year]: .2f}")
```

With this function I am able to analyze the performance of gold per year. And as said above it gives a baseline understanding how gold has performed.

Next we create a function showing the correlation of gold and silver

```python
def gold_silver_correlation (df):
    correlation = df ['gold'].corr(df ['silver'])
    print(f"Correlation: {correlation}")
    if correlation > 0.8:
        print("Positive correlation")
    elif correlation < 0.5:
        print("Weak correlation")
    elif correlation >= 0.5 and correlation <= 0.8:
        print("Moderate positive correlation")
    else:
        print("Negative correlation")
```

With this we can showcase the correlation of silver and gold and whether it had positive,negative,weak,and moderate coorrelation. 
We can do this by using the .corr() method 
Next create a function showcasing the best and worse months
A correlation of 0.8 is Positive
0.5 Weak
Between 0.5 and 0.8 is Moderately positive
and below 0.5 is Negative
0 means no correlations 

Gold and silver has 0.86 correlation meaning it has a positive and strong correlation. If gold rises, silver rises too. The main drive of golding going up is uncertainty on the economy such as recessions, inflation, and geopolitical/wars.

```python
def best_worst_months (df):
    month_names = {1: "January", 2: "February", 3: "March", 4: "April", 
               5: "May", 6: "June", 7: "July", 8: "August", 
               9: "September", 10: "October", 11: "November", 12: "December"}
               
    grouped_by_month = df.groupby('month') ['gold'].mean()
    best_month = grouped_by_month.idxmax()
    worst_month = grouped_by_month.idxmin()
    print(f"Best month: {month_names[best_month]} - ${grouped_by_month[best_month]:.2f}")
    print(f"Worst month: {month_names[worst_month]} - ${grouped_by_month[worst_month]:.2f}")
```

Start by making a dictionary of the months so they'll be easier to call. Use the groupby method to group the month and get the average of gold using mean
then use idx max to return the index label of the first occurence of the maximum value and viceversa with idxmin
Next creating the plot

Identifying which months historically perform best and worst 
for gold prices. This could be useful for timing investment decisions.

Febraury is historically the strongest month for gold while June is the weakest. Gold tends to perform better in the first quarter of the year. possibly because of the economic uncertainty at the start of each year.

```python
def plot_all(df):
    fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 8)) #indicate how many rows and columns
    
    # Chart 1
    axes[0].plot(df["date"], df["gold"], label="Gold")
    axes[0].plot(df["date"], df["silver"], label="Silver")
    axes[0].legend()
    axes[0].set_title("Gold vs Silver Price Over Time")
    axes[0].set_xlabel("Date")
    axes[0].set_ylabel("Price (USD)")
    
    # Chart 2
    grouped_by_year = df.groupby("year")["gold"].mean()
    axes[1].bar(grouped_by_year.index, grouped_by_year.values)
    axes[1].set_title("Yearly Average Gold Price")
    axes[1].set_xlabel("Year")
    axes[1].set_ylabel("Average Price (USD)")
    
    plt.tight_layout()
```

I chose subplot so it'll be easier to see the comparison of gold and silver throughout the years. On chart 1 it shows gold vs silver and on the second
it shows the Yearly gold price
Next is creating the function to save it to a file to summarize our findings

```python
def save_summary(df):
        highest = df['gold'].max()
        lowest = df['gold'].min()
        average = df['gold'].mean()
        correlation = df ['gold'].corr(df ['silver'])
        
        grouped_by_month = df.groupby('month') ['gold'].mean()
        best_month = grouped_by_month.idxmax()
        worst_month = grouped_by_month.idxmin()
        
        month_names = {1: "January", 2: "February", 3: "March", 4: "April", 
               5: "May", 6: "June", 7: "July", 8: "August", 
               9: "September", 10: "October", 11: "November", 12: "December"}
    
        with open("gold_summary.txt", "w") as file:
            file.write("=== Gold vs Silver Analysis ===\n\n")
            file.write(f"Highest Gold Price: ${highest:.2f}\n")
            file.write(f"Lowest Gold Price: ${lowest:.2f}\n")
            file.write(f"Average Gold Price: ${average:.2f}\n")
            file.write(f"Correlation: {correlation:.2f}\n")
            file.write(f"Best Month: {month_names[best_month]}\n")
            file.write(f"Worst Month: {month_names[worst_month]}\n")
```
